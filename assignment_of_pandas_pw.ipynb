```python
''' In pandas, a DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure with labeled axes (rows and columns). It is similar to a spreadsheet or SQL table, or a dictionary of Series objects.

A Series, on the other hand, is a one-dimensional array-like object containing a sequence of values and an associated array of data labels, called its index. It is essentially a single column of data.

Key Differences:
Dimensions:

Series: One-dimensional.
DataFrame: Two-dimensional.    '''
```

Pandas provides a rich set of functions to manipulate and analyze data in DataFrames. Here are some common functions along with examples of when you might use them:

Common Functions in Pandas DataFrame:
head() and tail():

Purpose: To view the first or last n rows of the DataFrame.
Example: When you want to quickly check the first few entries of a large DataFrame to understand its structure.

```python
df.head()  # By default, shows the first 5 rows
df.tail(3) # Shows the last 3 rows
```

info():

Purpose: To get a concise summary of the DataFrame, including the number of non-null entries and data types.
Example: Useful to understand the overall structure of the DataFrame and detect missing values.

```python
df.info()
```

describe():

Purpose: To generate descriptive statistics of the DataFrame's numeric columns.
Example: Useful for getting a quick summary of the central tendency, dispersion, and shape of the dataset’s distribution.

```python
df.describe()
```

drop():

Purpose: To remove rows or columns from the DataFrame.
Example: When you want to remove a column that is not needed for analysis.

```python
df.drop(columns=['UnnecessaryColumn'], inplace=True)
```

```python
df.loc[0:2, ['Name', 'Age']]  # Access by labels
df.iloc[0:2, [0, 1]]          # Access by positions
```

Example: When you need to apply a custom function to every element of a column.

```python
import pandas as pd

# Sample DataFrame
data = {
    'Product': ['Apple', 'Banana', 'Apple', 'Cherry', 'Banana', 'Apple'],
    'Quantity': [10, 20, 15, 10, 30, 10],
    'Price': [1.2, 0.5, 1.2, 2.5, 0.5, 1.2]
}

df = pd.DataFrame(data)

# View the first few rows
print("First few rows of the DataFrame:")
print(df.head())

# Group by product and calculate the total quantity sold for each product
print("\nTotal quantity sold for each product:")
total_quantity = df.groupby('Product')['Quantity'].sum()
print(total_quantity)

# Fill any missing values with 0 (if there were any)
df.fillna(0, inplace=True)
```

Q7. Create a DataFrame using multiple Series. Explain with an example.
To create a DataFrame using multiple Series, you can follow these steps:

Import the pandas library.
Create individual Series.
Combine the Series into a DataFrame.
# Here's an example:**

```python
import pandas as pd

# Create individual Series
names = pd.Series(['Alice', 'Bob', 'Claire'])
ages = pd.Series([25, 30, 27])
genders = pd.Series(['Female', 'Male', 'Female'])

# Combine the Series into a DataFrame
data = {
    'Name': names,
    'Age': ages,
    'Gender': genders
}

df = pd.DataFrame(data)

# Print the DataFrame
print(df)
```
